```python
t=np.arange(1,11)
y1=[12,23,45,65,10,3,14,16,32,33]
y2=[58,2,65,99,12,32,23,56,78,6]
y3=[45,3,23,65,91,73,82,19,61,45]

plt.stackplot(t,y1,y2,y3,colors=['r','k','b'])
plt.show()
```

```python
plt.subplot(2,2,1)
plt.plot(t,y1)
plt.scatter(t,y2)
plt.xlabel('time')
plt.ylabel('amplitude')

plt.subplot(2,2,4) #for (row,column,number)
plt.bar(t,y2,color='c')
plt.ylabel('sample')
plt.xlabel('data')
plt.show()
```

```python
plt.subplot(3,4,1)
plt.barh(t,y1,color='r')
plt.subplot(3,4,2)
plt.stackplot(t,y1,color='m')
plt.subplot(3,4,3)
plt.scatter(t,y3,color='g')
plt.subplot(3,4,4)
plt.plot(t,y1,color='c',linestyle='-')
plt.subplot(3,4,5)
plt.plot(t,y1,color='k',linestyle='-')
plt.subplot(3,4,6)
plt.bar(t,y1,color='c')
plt.bar(t,y2,color='c')
plt.subplot(3,4,7)
plt.barh(t,y1,color='r')
plt.scatter(t,y2,color='r')
plt.subplot(3,4,8)
plt.stackplot(t,y2,y3,color='b')
plt.subplot(3,4,9)
plt.bar(t,y1,color='m',linewidth='0')
plt.bar(t,y2,color='b',linewidth='2')
plt.subplot(3,4,10)
plt.scatter(t,y1,color='b')
plt.subplot(3,4,11)
plt.plot(t,y2,color='k')
plt.subplot(3,4,12)
plt.scatter(t,y2,color='g')
plt.scatter(t,y1,color='g')
plt.tight_layout()
plt.show()
```

```python
fig=plt.figure()
f1=plt.subplot2grid((3,4),(0,0),rowspan=1,colspan=3)
f2=plt.subplot2grid((3,4),(1,0),rowspan=2,colspan=2)
f3=plt.subplot2grid((3,4),(0,3),rowspan=3)
f4=plt.subplot2grid((3,4),(1,2),rowspan=2)

f1.barh(t,y2)
f2.plot(t,y1)
f3.bar(t,y3)
f4.scatter(t,y2)
plt.tight_layout()
plt.show()
```
